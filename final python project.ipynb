# Skopje Air Pollution

## Introduction

The city of Skopje is well known for its struggles with poor air quality, especially in the winter months. The main pollutants are CO, NO2, ozone, and particulate matter of different dimensions.

Monitoring stations for measuring the concentrations of each of the main pollutants have been installed in different locations around the city.

## Import Libraries
Before we begin, let us import the libraries that we will be using (Pandas, matplotlib and seaborn). Any additional libraries that we decide to import we will add in this cell.

To make our task more managable, we will only be looking at data for the time period of five years, from January 2012 to December 2016.

## Data cleaning and restructuring
Because the mobile column is from a mobile monitoring station located in the municipality of Butel and occasionally displaced, we can drop that column.

We can also rename the column time to datetime and create additional separate time and date columns. 

We can add a column in each of the dataframes indicating the type of pollutant and concatenate the dataframes.

```python
# Data cleaning and restructuring
df_CO = df_CO.drop(columns = ['Unnamed: 0.1', 'Unnamed: 0', 'Mobile'], axis = 1)
df_CO.rename(columns={'time': 'datetime'}, inplace=True)
df_CO['datetime'] = pd.to_datetime(df_CO.datetime)
df_CO['date'] = df_CO.datetime.dt.date
df_CO['time'] = df_CO.datetime.dt.time
df_CO['pollutant']= 'CO'
df_CO = df_CO.drop_duplicates()
```

```python
df_CO.info()
```

```python
df_NO2 = df_NO2.drop(columns = ['Unnamed: 0.1', 'Unnamed: 0', 'Mobile'], axis = 1)
df_NO2.rename(columns={'time': 'datetime'}, inplace=True)
df_NO2['datetime'] = pd.to_datetime(df_NO2.datetime)
df_NO2['date'] = df_NO2.datetime.dt.date
df_NO2['time'] = df_NO2.datetime.dt.time
df_NO2['pollutant']= 'NO2'
df_NO2 = df_NO2.drop_duplicates()
```

```python
df_O3 = df_O3.drop(columns = ['Unnamed: 0.1', 'Unnamed: 0', 'Mobile'], axis = 1)
df_O3.rename(columns={'time': 'datetime'}, inplace=True)
df_O3['datetime'] = pd.to_datetime(df_O3.datetime)
df_O3['date'] = df_O3.datetime.dt.date
df_O3['time'] = df_O3.datetime.dt.time
df_O3['pollutant']= 'O3'
df_O3 = df_O3.drop_duplicates()
```

# Question 1
What location in Skopje has the highest average pollution for each of the pollutants? .......... (2 points)

```python
average = new.groupby(['pollutant']).mean(numeric_only=True)
average.reset_index(inplace = True)
average
```

```python
for idx in average['pollutant']:
    redovi = pd.DataFrame(average[average['pollutant'] == idx])
    najzagaden = list(redovi.idxmax(axis = 1, numeric_only=True))
    ime = najzagaden[0]

    print(f'The highest average pollution in Skopje for {idx} pollutant is location: {ime}')
```

On how many days in each year, was the average value of PM10 meauserd accross locations worse than moderate?  .......... (3 points)

```python
PM10 = new[new['pollutant'] == 'PM10']
PM10 = PM10[['Centar', 'GaziBaba', 'Karpos', 'Lisice', 'Miladinovci', 'Rektorat','date', 'year']].groupby(['year', 'date']).mean()
average_daily_PM10 = PM10[PM10[['Centar', 'GaziBaba', 'Karpos', 'Lisice', 'Miladinovci', 'Rektorat']] > 154]
average_daily_PM10.reset_index(inplace=True)
avg_daily_PM10_worse = average_daily_PM10[['Centar', 'GaziBaba', 'Karpos', 'Lisice', 'Miladinovci', 'Rektorat','year']]
worse = avg_daily_PM10_worse.groupby(['year']).count()
worse
```

## Question 7
On how many days in each year, was the value of PM10 measured on at least one location worse than moderate? .......... (3 points)

```python
pm_10 = new[(new['pollutant']=='PM10')]
x = pm_10.set_index(['date','year'])
z = x[x.iloc[:,:6]> 154]
y = z.reset_index().groupby(['date','year']).count()
y['sum'] = y.sum(axis=1)
krajno = y[y['sum']>0]
krajno1 = krajno.groupby('year').count()
final = krajno1['sum']
final = pd.DataFrame(final)
final.rename(columns = {'sum':'Num_of_days'}, inplace = True)
final
```
